```
# Curso Data Visualization: criando gráficos com bibliotecas Python
#AULA 01 - Conhecendo a biblioteca Matplotlib
import pandas as pd
df = pd.read_csv('/content/imigrantes_canada.csv')
df
```

```
df.info()
```

```
df.set_index('País', inplace=True) #Setou o índex para que seja o país.
```

```
#variável anos
anos = list(map(str, range(1980, 2014))) #range não considera o último valor, por isso ele pega até 2013.
anos
```

```
brasil = df.loc['Brasil', anos] #Criar uma serie só com Brasil e os dados dos anos
```

```
brasil.head()
```

```
#colocar o nome das colunas, criando um dicionário
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()} #index pega os dados, tolist converte series em uma lista
dados_brasil = pd.DataFrame(brasil_dict)
```

```
dados_brasil #DF com os dados dos anos e quantidade de imigrações do Brasil.
```

```
import matplotlib.pyplot as plt
```

```
# gráfico por plotagem
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010']) #marcações do eixo "x", quero de 05 em 05 anos
#plt.yticks([500, 1000, 1500, 2000, 2500, 3000]) -> Faz marcações do eixo "y". Não precisa de '' pq não é strig como o ano. Já pode colocar o valor direto quando for inteiro.
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano') #Eixo x
plt.ylabel('Número de Imigrantes') #Eixo y
plt.show()
```

```
#atividade dentro da aula - plotar um gráfico de linhas utilizando a biblioteca matplotlib e assim visualizar o crescimento populacional de Águas de São Pedro ao longo do tempo.
df_aguas = pd.DataFrame({'ano': [1970, 1980, 1991, 2000, 2010, 2020],
                         'n_habitantes': [830, 1091, 1697, 1883, 2703, 3500]})
plt.plot(df_aguas['ano'], df_aguas['n_habitantes'])
```

```
#atividade dentro da aula - criar gráfico comparando as imigrações de Brasil e Argentina para o Canadá e responder às perguntas.
argentina = df.loc['Argentina', anos]
argentina_dict = {'ano': argentina.index.tolist(), 'imigrantes': argentina.values.tolist()}
dados_argentina = pd.DataFrame(argentina_dict)
dados_argentina
```

```
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'], label='Brasil')
plt.plot(dados_argentina['ano'], dados_argentina['imigrantes'], label='Argentina')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2015']) #marcações do eixo "x", quero de 05 em 05 anos
plt.title('Comparação da imigração do Brasil e da Argentina para o Canadá')
plt.xlabel('Ano') #Eixo x
plt.ylabel('Número de Imigrantes') #Eixo y
plt.legend()  # Legenda dos dados para saber qual país pertence a cada linha
plt.show()
```

```
#Há alguma tendência ou padrão comum nos dados dos dois países?
#R: Ambos os países apresentavam taxas parecidas, porém após 2005 aproximadamente ocorreu uma alteração significativa: enquanto o Brasil continuou crescendo, na Argentina o número de imigrações caiu consideravelmente.
#Quais são os períodos com maior número de imigrantes nos dois países?
#R: No Brasil é 2010 e na Argentina é 2003.
```

```
# tarefinha - criar quatro subplots dispostos em duas linhas e duas colunas,
# onde cada subplot representa uma loja diferente. Nesse desafio, cada subplot
# deve apresentar um gráfico de linhas que mostre a variação do número de
# vendas ao longo dos meses do ano.

lojas = ['A', 'B', 'C', 'D']
vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}

df_vendas = pd.DataFrame(vendas_2022, index=lojas)
df_vendas.index.name = 'loja'
df_vendas
```

```
fig, axs = plt.subplots(2,2, figsize=(14,8))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Evolução das vendas de 2022')

axs[0,0].plot(df_vendas.loc['A'])
#axs[0, 0].plot(df.columns, df.loc['A'], label='Loja A')
#axs[0, 0].plot(df.loc['A', df.columns])
axs[0,0].set_title('Vendas loja A')

axs[0,1].plot(df_vendas.loc['B'])
axs[0,1].set_title('Vendas loja B')

axs[1,0].plot(df_vendas.loc['C'])
axs[1,0].set_title('Vendas loja C')

axs[1,1].plot(df_vendas.loc['D'])
axs[1,1].set_title('Vendas loja D')

for ax in axs.flat:
  ax.set_xlabel('Meses')
  ax.set_ylabel('Número de vendas')
  ax.grid()

ymin = 0
ymax = 500
for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

#contribuição de visualização do fórum: Colcoar os valores, rótulos de dados
  for i, valor in enumerate(df_vendas.loc[ax.get_title().split()[-1]]):
        ax.annotate(f'{valor}', (i, valor), textcoords="offset points", xytext=(0,5), ha='center')

plt.show()
```

```
def gerar_grafico_paleta(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax = sns.barplot(data=top_10, y=top_10.index, hue=top_10.index, x='Total', orient='h', palette=palette)
  ax.set_title('Países com maior imigração para o Canadá\n1908 a 2013', loc='left', fontsize=18)
  ax.set_xlabel('Número de imigrantes', fontsize=14)
  ax.set_ylabel('')
  plt.show()
```
